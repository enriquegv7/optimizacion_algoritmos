## Torres de Hanoi con Divide y vencerás

Resuelve el problema de las [Torres de Hanoi](https://es.wikipedia.org/wiki/Torres_de_Han%C3%B3i) para un número arbitrario de fichas utilizando la técnica de divide y vencerás. 

```python
# Torres de Hanoi
def hanoi(n, origen, destino, auxiliar):
    if n == 1:
        print("Mover disco 1 desde {} hasta {}".format(origen, destino))
        return
    hanoi(n-1, origen, auxiliar, destino)
    print("Mover disco {} desde {} hasta {}".format(n, origen, destino))
    hanoi(n-1, auxiliar, destino, origen)


def resolver_hanoi(num_discos):
    print("Pasos para resolver las Torres de Hanoi con {} discos:".format(num_discos))
    hanoi(num_discos, 'Torre 1', 'Torre 3', 'Torre 2')

# Ejemplo de uso
resolver_hanoi(3)
resolver_hanoi(5)
```

## Sucesión de Fibonacci

Cálcula el n-ésimo término de la [Sucesión de Fibonacci](https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci) mediante un algoritmo recursivo y otro iterativo. Representa gráficamente cómo crece el tiempo de cómputo en función del número de términos para ambos algoritmos. 

```python
#Sucesión_de_Fibonacciç
import time
import matplotlib.pyplot as plt

# Recursivo 
def fibonacci_recursivo(n):
    if n <= 1:
        return n
    else:
        return fibonacci_recursivo(n-1) + fibonacci_recursivo(n-2)

#Iterativo
def fibonacci_iterativo(n):
    if n <= 1:
        return n
    else:
        fib_ant = 0
        fib_actual = 1
        for _ in range(2, n+1):
            fib_siguiente = fib_ant + fib_actual
            fib_ant, fib_actual = fib_actual, fib_siguiente
        return fib_actual
    
#Ejemplo 1
n = 10
resultado = fibonacci_iterativo(n)
print("Iterativo - El {}-ésimo término de la sucesión de Fibonacci es: {}".format(n, resultado))
resultado = fibonacci_recursivo(n)
print("Recursivo - El {}-ésimo término de la sucesión de Fibonacci es: {}".format(n, resultado))

#Ejemplo2
n = 20
resultado = fibonacci_iterativo(n)
print("Iterativo - El {}-ésimo término de la sucesión de Fibonacci es: {}".format(n, resultado))
resultado = fibonacci_recursivo(n)
print("Recursivo - El {}-ésimo término de la sucesión de Fibonacci es: {}".format(n, resultado))



valores_n = list(range(1, 35))  # Considerando hasta n = 34 para evitar tiempos de ejecución muy largos
tiempos_recursivo = []
tiempos_iterativo = []

for n in valores_n:
    inicio = time.perf_counter()
    fibonacci_recursivo(n)
    fin = time.perf_counter()
    tiempos_recursivo.append(fin - inicio)
    
    inicio = time.perf_counter()
    fibonacci_iterativo(n)
    fin = time.perf_counter()
    tiempos_iterativo.append(fin - inicio)

# Graficar los tiempos de ejecución en función de n
plt.plot(valores_n, tiempos_recursivo, label='Recursivo', marker='o', linestyle='-')
plt.plot(valores_n,tiempos_iterativo, label='Iterativo', marker='o', linestyle='-')
plt.title('Tiempo de ejecución de los algoritmos de Fibonacci')
plt.xlabel('Número de términos (n)')
plt.ylabel('Tiempo de ejecución (segundos)')
plt.legend()
plt.grid(True)
plt.show()
```

## Devolución de cambio por técnica voraz

Resuelve el [Problema del Cambio de Moneda](https://es.wikipedia.org/wiki/Problema_de_cambio_de_monedas) utilizando una técnica voraz.

```python
# Problema del cambio de moneda
def cambio_de_moneda(monedas, cantidad):
    monedas = sorted(monedas, reverse=True)
    num_monedas = 0
    cambio = []

    for moneda in monedas:
        num = cantidad // moneda
        cambio.extend([moneda] * num)
        cantidad -= moneda * num
        num_monedas += num
        if cantidad == 0:
            break

    return cambio, num_monedas

# Ejemplo1
monedas = [40,20, 10, 5, 1]
cantidad = 55 

cambio, num_monedas = cambio_de_moneda(monedas, cantidad)
print("Cambio obtenido:", cambio)
print("Número total de monedas utilizadas:", num_monedas)


# Ejemplo2
monedas = [50, 20, 10, 5, 1]
cantidad = 80 

cambio, num_monedas = cambio_de_moneda(monedas, cantidad)
print("Cambio obtenido:", cambio)
print("Número total de monedas utilizadas:", num_monedas)
```

## N-Reinas por técnica de vuelta atrás
Resuelve el [Problema de las N-Reinas](https://es.wikipedia.org/wiki/Problema_de_las_ocho_reinas) en un tablero de dimensión N mediante la técnica de la vuelta atrás (backtraking).

```python
# Problema de las N-Reinas
def es_seguro(tablero, fila, columna, N):
    for i in range(fila):
        if tablero[i][columna] == 1:
            return False
    
    for i, j in zip(range(fila, -1, -1), range(columna, -1, -1)):
        if tablero[i][j] == 1:
            return False
    
    for i, j in zip(range(fila, -1, -1), range(columna, N)):
        if tablero[i][j] == 1:
            return False
    
    return True

def resolver_n_reinas_util(tablero, fila, N):
    if fila == N:
        return True
    
    for columna in range(N):
        if es_seguro(tablero, fila, columna, N):
            tablero[fila][columna] = 1
            if resolver_n_reinas_util(tablero, fila + 1, N):
                return True
            
            tablero[fila][columna] = 0
            
    return False

def resolver_n_reinas(N):
    tablero = [[0] * N for _ in range(N)]
    
    if resolver_n_reinas_util(tablero, 0, N):
        return tablero
    else:
        return None

def imprimir_tablero(tablero):
    if tablero is None:
        print("No se encontró una solución.")
    else:
        for fila in tablero:
            print(' '.join(map(str, fila)))

# Ejemplo1
N = 4 
solucion = resolver_n_reinas(N)
imprimir_tablero(solucion)
print("------------------------------------------")

# Ejemplo2
N = 20
solucion = resolver_n_reinas(N)
imprimir_tablero(solucion)
```
